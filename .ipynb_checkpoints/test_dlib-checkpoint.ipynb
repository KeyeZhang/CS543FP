```python
import cv2 as cv
import matplotlib.pyplot as plt
import time
from PIL import Image
import numpy as np 
from skimage.measure import block_reduce
import pylab
import os
```

```python
test1 = cv.imread('./0_Parade_Parade_0_1019.jpg')
gray1 = cv.cvtColor(test1, cv.COLOR_BGR2GRAY)
# haar_face_cascade = cv.CascadeClassifier('./haarcascades/haarcascades/haarcascade_frontalface_default.xml')

t = time.time()
lbp_face_classifier = cv.CascadeClassifier('./lbpcascades/lbpcascades/lbpcascade_frontalface_improved.xml')
# plt.imshow(gray1, cmap='gray')
# plt.show()

def detect_faces(f_cascade, colored_img, scaleFactor = 1.005):
 #just making a copy of image passed, so that passed image is not changed 
    img_copy = colored_img.copy()
 #convert the test image to gray image as opencv face detector expects gray images
    gray = cv.cvtColor(img_copy, cv.COLOR_BGR2GRAY)
 #let's detect multiscale (some images may be closer to camera than others) images
    faces = f_cascade.detectMultiScale(gray, scaleFactor=scaleFactor, minNeighbors=5)
 #go over list of faces and draw them as rectangles on original colored img
    for (x, y, w, h) in faces:
        cv.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)
    return img_copy
# detected_faces()
faces = detect_faces(lbp_face_classifier, test1)

t = time.time() - t
plt.subplot(121)
plt.imshow(cv.cvtColor(faces, cv.COLOR_BGR2RGB))
plt.subplot(122)
plt.imshow(cv.cvtColor(test1, cv.COLOR_BGR2RGB))
plt.show()
t
```
