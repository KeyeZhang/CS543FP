```python
#import required libraries 
#import OpenCV library
import cv2
#import matplotlib library
import matplotlib.pyplot as plt
#importing time library for speed comparisons of both classifiers
import time 
%matplotlib inline
```

```python
#implemetation choice
own_classifier = False  #false / true
type_classifier = 2 #1: haar / 2: lbp
```

```python
# #convert the default OpenCV BGR space to RGB space
#load test iamge
test1 = cv2.imread('./WIDER_train/images/16--Award_Ceremony/16_Award_Ceremony_Awards_Ceremony_16_41.jpg')
# plt.imshow(test1, cmap='gray')
#convert the test image to gray image as opencv face detector expects gray images 
gray_img = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)

#if you have matplotlib installed then  
plt.imshow(gray_img)  

time1 = time.time();
#classifier detection choice
if type_classifier == 1: 
    if own_classifier == True:
        face_classifier = cv2.CascadeClassifier('classifier/haarcascade_cs543.xml');
    else:
        face_classifier = cv2.CascadeClassifier('classifier/haarcascade_frontalface_alt2.xml');
elif type_classifier == 2:
    if own_classifier == True:
        face_classifier = cv2.CascadeClassifier('classifier/lbpcascade_cs543.xml');
    else:
        face_classifier = cv2.CascadeClassifier('classifier/lbpcascade_frontalface.xml');
else:
    #do nothing
    print("do not have this method");


#faces = face_classifier.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5); 
faces = face_classifier.detectMultiScale(gray_img, minNeighbors=5); 
    
#print the number of faces found 
print('Total faces numbers: ', len(faces))

time2 = time.time();
totalTime = time2 - time1;
print('Total running time: ', totalTime);
#There are several parameters:
#For example, the scaleFactor make up for the difference of faces size due to the distance to the camera.
#minNeighbors is used when the algorithm is scanning each pixel with a window, and the minNeighbor decides at least 
#how many objects it should found to delcare that the object is a face.

for (x, y, w, h) in faces:     
         cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 255), 5)
        
#convert image to RGB and show image 
plt.imshow(cv2.cvtColor(test1, cv2.COLOR_BGR2RGB))
```
